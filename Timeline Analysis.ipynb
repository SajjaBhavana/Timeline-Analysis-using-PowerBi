```python
data_set['date'] = pd.to_datetime(data_set['date'])
data_set.head()
```

```python
data_set['Month-Year']= data_set['date'].dt.strftime('%b-%y')
data_set.head()
```

```python
data_set['continent'].fillna(data_set['location'],inplace=True)
data_set['total_cases'].fillna(0,inplace=True)
data_set['total_deaths'].fillna(0,inplace=True)
data_set['new_cases'].fillna(0,inplace=True)
data_set['new_deaths'].fillna(0,inplace=True)
data_set['reproduction_rate'].fillna(round(data_set['reproduction_rate'].mean(),2),inplace=True)
data_set['stringency_index'].fillna(round(data_set['stringency_index'].mean(),2),inplace=True)
data_set['population'].fillna(round(data_set['population'].mean(),2),inplace=True)
data_set['aged_65_older'].fillna(round(data_set['aged_65_older'].mean(),2),inplace=True)
data_set['aged_70_older'].fillna(round(data_set['aged_70_older'].mean(),2),inplace=True)
data_set['cardiovasc_death_rate'].fillna(round(data_set['cardiovasc_death_rate'].mean(),2),inplace=True)
data_set['diabetes_prevalence'].fillna(round(data_set['diabetes_prevalence'].mean(),2),inplace=True)
```
